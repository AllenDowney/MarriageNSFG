```python
def add_newlines():
    filename = '1988FemRespData.dat'
    fin = open(filename, 'r')
    fout = open('1988FemRespDataLines.dat', 'w')

    for i in range(8450):
        line = fin.read(3553)
        fout.write(line + '\n')
    
    fout.close()
```

```python
filename = '1988FemRespDataLines.dat.gz'
names = ['finalwgt', 'ageint', 'cmmarrhx', 'cmintvw', 'cmbirth']
colspecs = [(2568-1, 2574),
            (36-1, 37),
            (1538-1, 1542),
            (12-1, 16),
            (26-1, 30),
            ]
df = pandas.read_fwf(filename,
                     colspecs=colspecs, 
                     names=names,
                     header=None,
                     compression='gzip')
```
